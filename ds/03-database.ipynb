```python
#AS - 생략 가능

query = """
SELECT Name AS name FROM 'transaction'
"""
pd.read_sql(query,connect)
```

```python
query = """
SELECT AVG(amount) FROM 'transaction'

"""
pd.read_sql(query,connect)
```

```python
query = """
SELECT AVG(amount), * FROM 'transaction'

"""
pd.read_sql(query,connect)
```

```python
# aggfunc = aggregate function = 집계 함수 
# HAVING = WHERE for group by column
# <> : != 
query = """
SELECT date, AVG(amount)*100 AS avg, SUM(amount) FROM 'transaction'
GROUP BY date
HAVING date !='2017-01-01'

"""
pd.read_sql(query,connect)
```

```python
# DISTINCT = unique() in pandas 
query = """
SELECT DISTINCT(amount) FROM 'transaction'
"""

pd.read_sql(query,connect)
```

## WHERE

```python
query = """
SELECT * FROM 'transaction'
WHERE (amount > 500) AND (amount < 900)
"""

pd.read_sql(query,connect)
```

```python
# between is inclusive of end points 

query = """
SELECT * FROM 'transaction'
WHERE (amount BETWEEN 500 AND 900)
    OR (amount > 100 AND amount < 300)
"""

pd.read_sql(query,connect)
```

```python
query = """
SELECT * FROM 'transaction'
WHERE result != 'confirmed'
"""

pd.read_sql(query,connect)
```

```python
query = """
SELECT * FROM 'transaction'
WHERE product NOT IN ("apple", "banana")
"""

pd.read_sql(query,connect)
# 굳이 or 을 여러번 쓸 필요 없음! 
```

```python
query = """
SELECT * FROM 'transaction'
WHERE product LIKE "a%"
"""

pd.read_sql(query,connect)
```

```python
# p가 두번째인거 찾기
query = """
SELECT * FROM 'transaction'
WHERE product LIKE "_p%"
"""

pd.read_sql(query,connect)
```

```python
# a나 k가 위치에 상관없이 들어간거 찾기
query = """
SELECT * FROM 'transaction'
WHERE product LIKE "%a%" 
    OR product LIKE "%k%"
"""

pd.read_sql(query,connect)
```

```python
# a가 들어가면서 최소 2글자인거 찾기
query = """
SELECT * FROM 'transaction'
WHERE product LIKE "%a_%" 
"""

pd.read_sql(query,connect)
```

## Join

```python
query = """
SELECT * FROM 'transaction'
"""

pd.read_sql(query, connect)
```

```python
query = """
SELECT * FROM 'user'
"""

pd.read_sql(query, connect)
```

```python
# default is INNER JOIN

query = """
SELECT * FROM 'transaction'
JOIN 'user' ON 'transaction'.Name = 'user'.Name
"""

pd.read_sql(query, connect)
```

```python
# outer랑 right join은 sqlite3에 없음

query = """
SELECT * FROM 'transaction'
LEFT JOIN 'user' ON 'transaction'.Name = 'user'.Name
"""

pd.read_sql(query, connect)
```

```python
query = """
SELECT * FROM 'transaction' A
JOIN 'user' B ON A.Name = B.Name
"""

pd.read_sql(query, connect)
```

```python
# 순서는 limit이 마지막. 그 다음이 ORDER BY

query = """
SELECT A.date DATE, B.* FROM 'transaction' A
JOIN 'user' B ON A.Name = B.Name
WHERE AGE > 20 
    AND Sex = 'male'
ORDER BY DATE
"""

pd.read_sql(query, connect)
```

```python
query = """
SELECT A.date DATE, B.*, C.* FROM 'transaction' A
INNER JOIN 'user' B ON A.Name = B.Name
INNER JOIN 'product' C ON A.product = C.product
WHERE C.price > 4
ORDER BY DATE
"""

pd.read_sql(query, connect)
```

# Subquery 

```python
query = """
SELECT * FROM 
    (SELECT * FROM 'transaction'
    WHERE amount > 500)
"""

pd.read_sql(query,connect)
```

```python
# product가 기준이니까 amount 가 500인 apple도 나옴. apple이 700도 있으니까. 
query = """
SELECT * FROM 'transaction'
WHERE product IN 
    (SELECT product FROM 'transaction'
    where amount > 500)
"""

pd.read_sql(query,connect)
```
