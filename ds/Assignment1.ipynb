```python
import pandas as pd
pd.options.display.max_columns = 100
raw_log = pd.read_csv("data/kmong-conversion.csv")
print(raw_log.shape)
raw_log.head(3)
```

# 1)

```python
raw_log['userid'] = raw_log['canonicaldeviceuuid']
cols = ['canonicaldeviceuuid','userid']
raw_log[cols].head()
```

# 2)

```python
raw_log['eventdatetime'] = pd.to_datetime(raw_log['eventdatetime'])

raw_log['eventdatetime_year'] = raw_log['eventdatetime'].dt.year
raw_log['eventdatetime_month'] = raw_log['eventdatetime'].dt.month
raw_log['eventdatetime_day'] = raw_log['eventdatetime'].dt.day
raw_log['eventdatetime_hour'] = raw_log['eventdatetime'].dt.hour
raw_log['eventdatetime_minute'] = raw_log['eventdatetime'].dt.minute
raw_log['eventdatetime_second'] = raw_log['eventdatetime'].dt.second

cols= ['eventdatetime','eventdatetime_year','eventdatetime_month','eventdatetime_day','eventdatetime_hour','eventdatetime_minute','eventdatetime_second']

raw_log[cols].head()
```

# 3) 

```python
def get_osversion(osversion):
    return osversion.replace("Android","").replace("iOS","")

def get_ostype(osversion):
    if "iOS" in osversion:
        return "iOS"
    else:
        return "Android"
    
raw_log['ostype(clean)'] = raw_log['osversion'].apply(get_ostype)
raw_log['osversion(clean)'] = raw_log['osversion'].apply(get_osversion)

cols = ['osversion','ostype(clean)','osversion(clean)']

raw_log[cols].head()
```

# 4)

```python
raw_log['devicemanufacturer'].unique()
```

```python
def clean_devicemanufacturer(devicemanufacturer):
    if devicemanufacturer == 'samsung':
        return 'Samsung'
    elif devicemanufacturer == 'Apple':
        return 'Apple'
    elif "LG" in devicemanufacturer:
        return "LG"
    else:
        return "others"
    
raw_log['devicemanufacturer(clean)'] = raw_log['devicemanufacturer'].apply(clean_devicemanufacturer)

cols = ['devicemanufacturer','devicemanufacturer(clean)']
raw_log[cols].head()
```

```python
raw_log['devicemanufacturer(clean)'].value_counts()
```

# 5)

```python
raw_log['channel'].unique()
```

```python
def clean_channel(channel):
    while pd.isna(channel) == False:
        if "naver" in channel:
            return "naver"
        elif "WEB" in channel:
            return "web"
        elif "google" in channel:
            return "google"
        elif "apple" in channel:
            return "apple"
        elif "daum" in channel:
            return "daum"
        else:
            return channel
            
raw_log['channel(clean)'] = raw_log['channel'].apply(clean_channel)
# cols = ['channel','channel(clean)']
raw_log['channel(clean)'].value_counts()
```

# 6)

```python
def view_category(inappeventcategory):
    while pd.isna(inappeventcategory) == False: 
        if "_" in inappeventcategory:
            arr = inappeventcategory.split("_")
            return arr[0]
        elif "." in inappeventcategory:
            arr = inappeventcategory.split(".")
            return arr[0]
        
def view_id(inappeventcategory):
    while pd.isna(inappeventcategory) == False: 
        if "." in inappeventcategory:
            arr = inappeventcategory.split(".")
            return arr[0]
        
def view_action(inappeventcategory):
    while pd.isna(inappeventcategory) == False: 
        if "." in inappeventcategory:
            arr = inappeventcategory.split(".")
            return arr[1]

raw_log['viewcategory'] = raw_log['inappeventcategory'].apply(view_category)
raw_log['viewid'] = raw_log['inappeventcategory'].apply(view_id)
raw_log['viewaction'] = raw_log['inappeventcategory'].apply(view_action)

cols = ['inappeventcategory','viewcategory','viewid','viewaction']

raw_log[cols].head()
```
