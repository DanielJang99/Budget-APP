```python
# user['age'] = user['age'].replace(np.nan,-1)
def make_generation(age):
    if age == -1:
        return ''
    elif age // 10 >= 4:
        return "30 "
    elif age // 10 == 1:
        return "10"
    elif age % 10 < 3:
        return str(age // 10 * 10) + f" "
    elif age % 10 <= 6:
        return str(age // 10 * 10) + f" "
    else:
        return str(age // 10 * 10) + f" "
    
def to_int(age):
    return int(age)

    
user['age'] = user['age'].apply(to_int)
user[''] = user['age'].apply(make_generation)
user
```

```python
# set()  user_id        .
user_id_set = set(order['user_id'])
#    user_number_order  .

user_number_order = len(user_id_set)

# log    user_number_log  .

user_number_log = len(set(log['user_id']))

#   user_duplicated  .

user_duplicated = set(order['user_id']) & set(log['user_id'])

print('      :',user_number_order, end ='\n')
print('  log    :',user_number_log,  end ='\n')
print('  :', len(user_duplicated) , end ='\n')
```

```python
# order     order copy . 
order_copy = order.copy()
```

```python
#   data indexing. user_duplicated . 
# isin  indexing        .
order_copy = order_copy[order_copy['user_id'].isin(user_duplicated)]

#  event_origin  shop_id  .
order_copy['event_origin'] = order_copy['shop_id']

#  event_name  'purchase' .
order_copy['event_name'] = 'purchase'

# event_goods_id  good_id  .
order_copy['event_goods_id'] = order_copy['goods_id']

#  columns .
order_copy = order_copy[['timestamp', 'n_user_id', 'user_id','event_origin',
         'event_name', 'event_goods_id', 'price']]

order_copy.head()
```

```python
log_order = pd.concat([log, order_copy])
log_order.head()
```

```python
def ReturnPurchase(data):
    if pd.isnull(data):
        return False
    else:
        return True

log_order['purchase']=log_order['price'].apply(ReturnPurchase)
log_order
```

```python
cols=['event_goods_id', 'event_name', 'event_origin', 'event_shop_id', 'n_user_id', 'price', 'timestamp', 'user_id', 'purchase']
log_order = log_order.sort_values(by=["user_id","timestamp"]).reset_index()
log_order.head()
```

# 13)

```python
log_order['timestamp_after'] = log_order.groupby(['n_user_id'])['timestamp'].shift(-1)

log_order.head()
```

```python
log_order['log_duration'] =  log_order['timestamp_after'] - log_order['timestamp']
log_order['log_duration']= log_order['log_duration'].replace(np.datetime64('NaT'), np.timedelta64(0,'D'))
log_order[['timestamp','timestamp_after','log_duration']].head()
```

```python
def return_seconds(data):
    return data.total_seconds()
log_order['log_duration'] = log_order['log_duration'].map(return_seconds)
log_order['log_duration'] = log_order['log_duration'].astype(float)
log_order['log_duration']
```

# 14)

```python
def IsOut(data):
    if data == 0 or data > 2400:
        return True
    else:
        return False

log_order['is_out'] = log_order['log_duration'].apply(IsOut)
log_order[['n_user_id','timestamp', 'timestamp_after', 'log_duration', 'is_out']].head()
```

```python
def EditLogCycle(data):
    if data == 0 or data > 2400:
        return 0
    else:
        return data

log_order['log_duration'] = log_order['log_duration'].apply(EditLogCycle)
log_order[['n_user_id','timestamp', 'timestamp_after', 'log_duration', 'is_out']].head()
```

```python
log_order['cycle_idx_unique'] = (log_order['is_out']
                                 .cumsum()
                                 .shift(1)
                                 .fillna(0)
                                 .astype(int)
                                )

log_order[['n_user_id','timestamp', 'timestamp_after', 'log_duration', 'is_out', 'cycle_idx_unique']].head(15)
```

```python
log_order['cycle_idx_daily'] = (log_order.groupby('n_user_id')['is_out']
                                           .cumsum()
                                           .shift(1)
                                           .fillna(0)
                                           .astype(int)
                                            )

# user  cycle_idx_daily 1    0 .
head_index = log_order.groupby('n_user_id')['cycle_idx_daily'].head(1).index
log_order.loc[head_index, 'cycle_idx_daily'] = 0

log_order[['n_user_id','timestamp', 'timestamp_after', 'log_duration', 'is_out','cycle_idx_daily']].head(30)
```

## 15.2)

```python
cycle_purchase = log_order['cycle_idx_unique'][log_order['purchase'] == True]
cycle_purchase
```

```python
data_purchase = log_order.loc[log_order['cycle_idx_unique'].isin(cycle_purchase)]
data_npurchase = log_order.loc[~log_order['cycle_idx_unique'].isin(cycle_purchase)]
```

```python
purchase_cycle_time = data_purchase.groupby(['n_user_id','cycle_idx_unique'])['log_duration'].sum().reset_index()
purchase_cycle_time = purchase_cycle_time.rename(columns = {'log_duration': 'cycle_duration'})
purchase_cycle_time.head(10)
```

```python
nonpurchase_cycle_time = data_npurchase.groupby(['n_user_id','cycle_idx_unique'])['log_duration'].sum().reset_index()
nonpurchase_cycle_time = nonpurchase_cycle_time.rename(columns = {'log_duration': 'cycle_duration'})
nonpurchase_cycle_time.head(10)
```

```python
figure, (ax1, ax2) = plt.subplots(nrows=2, ncols=1)

sns.boxplot(purchase_cycle_time['cycle_duration'], whis =2, ax= ax2)
sns.boxplot(nonpurchase_cycle_time['cycle_duration'], ax = ax1)
```

```python
mean1 = purchase_cycle_time['cycle_duration'].mean()
mean2 = nonpurchase_cycle_time['cycle_duration'].mean()
print(" cycle   :", mean1)
print(" cycle   :", mean2)
```
