```python
cols =['Name','Age(clean)','Height(clean)','Initial Weight','Lowest Weight','Target Weight','Status']
weight_history = data[cols].copy()
weight_history.head()
```

# 6)

```python
weight_history["BMI"] = (weight_history['Initial Weight']/((weight_history['Height(clean)']/100))**2)
weight_history['Weight Loss(goal)'] = (weight_history['Initial Weight']-weight_history['Target Weight'])
weight_history['Weight Loss(current)'] = (weight_history['Initial Weight']-weight_history['Lowest Weight'])

cols = ['Name','Weight Loss(goal)','Weight Loss(current)','BMI']
weight_history[cols].head()
```

# 7)

```python
def InvalidUser(frame): 
    if frame['Status'] == "completed":
        if pd.isna(frame['Age(clean)']):
            return True  
        elif pd.isna(frame['Height(clean)']):
            return True
        elif pd.isna(frame['Initial Weight']):
            return True
        elif pd.isna(frame['Lowest Weight']):
            return True
        elif pd.isna(frame['Target Weight']):
            return True
        elif frame['Height(clean)'] < 140 or frame['Height(clean)'] > 200:
            return True
        elif frame['BMI'] < 18.5 or frame['BMI'] > 30.0:
            return True
        elif frame['Weight Loss(goal)'] < 0:
            return True
        else:
            return False
    else:
        return False

weight_history['Invalid'] = weight_history.apply(InvalidUser, axis=1).copy()
weight_history['Invalid'].value_counts()
```

# 8)

```python
def VIPUser(frame):
    if frame['Status'] != "completed":
        return False
    elif pd.isna(frame['Weight Loss(goal)']):
        return False
    elif pd.isna(frame['Weight Loss(current)']):
        return False
    elif pd.isna(frame['BMI']):
        return False
    elif frame['Weight Loss(current)'] < 10:
        return False
    elif frame['BMI'] < 30:
        return False
    elif frame['Weight Loss(current)'] < frame['Weight Loss(goal)']:
        return False
    else:
        return True
    
weight_history['VIP'] = weight_history.apply(VIPUser, axis=1).copy()
weight_history['VIP'].value_counts()
```

# 10)

```python
def SortAge(data):
    if data <= 17.0:
        return '00 ~ 17'
    if data >= 18.0 and data <= 24.0:
        return '18 ~ 24'
    if data >= 25.0 and data <= 35.0:
        return '25 ~ 35'
    if data >= 36.0 and data <= 44.0:
        return '36 ~ 44'
    if data >= 45.0 and data <= 54.0:
        return '45 ~ 54'
    if data >= 55.0:
        return '55 ~ 99'

data['Age(group)'] = data['Age(clean)'].apply(SortAge)
data['Age(group)'].value_counts()
```

```python
table1 = data.pivot_table(index = ['Gender(clean)','Age(group)'], columns =['Status'],fill_value=0,aggfunc=['size'])
table1
```

```python
table2 = data.pivot_table(index = ['Gender(clean)'], columns=['Age(group)'],values=['Status'],aggfunc='count',fill_value=0).stack('Age(group)')
table2
```

```python
table= pd.merge(table1['size'],table2['Status'],on=['Gender(clean)','Age(group)'],how='left')
table
```

```python
table['conversion'] = table['completed']/table['Status']
table
```

```python
data['Purchased At'].dt.hour
```

```python
def GetPurchasedHour(data):
    return data.hour

data['Purchased At(hour)'] = data['Purchased At'].apply(GetPurchasedHour).copy()
```

```python
table1 = data.pivot_table(index=['Purchased At(hour)'], columns =['Status'],fill_value=0,aggfunc=['size'])
```

```python
table2 = data.pivot_table(index = ['Purchased At(hour)'],values=['Status'],aggfunc='count',fill_value=0)
```

```python
table = pd.merge(table1['size'],table2, on=['Purchased At(hour)'], how= "left")
```

```python
table['conversion'] = table['completed']/table['Status']
table.head()
```

```python
def GetPurchasedDay(data):
    if data.weekday() == 0:
        return "Monday"
    elif data.weekday() == 1:
        return "Tuesday"
    elif data.weekday() == 2:
        return "Wednesday"
    elif data.weekday() == 3:
        return "Thursday"
    elif data.weekday() == 4:
        return "Friday"
    elif data.weekday() == 5:
        return "Saturday"
    else:
        return "Sunday"
    

data['Purchased At(weekday)'] = data['Purchased At'].apply(GetPurchasedDay).copy()
```

```python
table1 = data.pivot_table(index=['Purchased At(weekday)'], columns =['Status'],fill_value=0,aggfunc=['size'])
table2 = data.pivot_table(index = ['Purchased At(weekday)'],values=['Status'],aggfunc='count',fill_value=0)
table = pd.merge(table1['size'],table2, on=['Purchased At(weekday)'], how= "left")
table['conversion'] = table['completed']/table['Status']
idx = ["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]
table.reindex(idx)
```
