### violinplot

```python
sns.violinplot(data['허리둘레'])
# 분포를 대칭으로 보여줌 
```

```python
plt.figure(figsize=(10,5))

sns.violinplot(data=data, x='연령대코드(5세단위)', y='허리둘레', orient='v')
```

```python
plt.figure(figsize=(10,10))

sns.violinplot(data=data, x='허리둘레', y='연령대코드(5세단위)', orient='h')
# orient를 h로 바꾸면 기준도 바뀜. 
```

```python
plt.figure(figsize=(10,10))

sns.violinplot(data=data, x='허리둘레', y='연령대코드(5세단위)', orient='h', hue = '성별코드', split = True)
# split으로 반반 나눔. 
```

### barplot 

```python
sns.barplot(data=data, x='연령대코드(5세단위)', y='허리둘레')
# 검은색 라인은 95% 신뢰구간 
```

```python
sns.barplot(data=data, x='연령대코드(5세단위)', y='허리둘레', ci=95, hue='성별코드')
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=data, x='연령대코드(5세단위)', y='허리둘레', estimator=lambda group_series: group_series.mean())
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=data, x='연령대코드(5세단위)', y='허리둘레', estimator=lambda group_series: group_series.max() - group_series.min(), ci=0)
# range 구하기 
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=data, x='연령대코드(5세단위)', y='허리둘레', estimator=lambda x: x.mean())
```

### boxplot

```python
# outlier을 확인할때 사용
# box는 IQR; also called the midspread, middle 50%, being equal to the difference between 75th and 25th percentile

plt.figure(figsize=(10,5))

sns.boxplot(data['허리둘레'], whis = 2)
# whis로 fence를 조정 
```

```python
plt.figure(figsize=(10,5))

sns.boxplot(data=data, x='연령대코드(5세단위)',y='허리둘레', hue = '성별코드')
```

```python
a = data['연령대코드(5세단위)'].unique()

a.sort()

a
```

```python
plt.figure(figsize=(10,5))

sns.boxplot(data=data, x='연령대코드(5세단위)',y='허리둘레', hue = '성별코드', order = a)
```

### lineplot

```python
#pointplot 
plt.figure(figsize=(10,5))
sns.lineplot(data=data, x='허리둘레', y='총콜레스테롤')
# 하늘색 범위는 신뢰구간 
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(data=data, x='허리둘레', y='총콜레스테롤', err_style='bars')
# err_style로 신뢰구간을 숨기기. 
# Marker로 
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(data=data, x='허리둘레', y='총콜레스테롤', err_style='band', markers=True, hue='성별코드', style='성별코드')
# makers로 꺽이는 점마다 표시를 해둠. 꼭 style을 표시해야됌. 
```

### scatterplot

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=data, x='수축기혈압', y='총콜레스테롤', hue='허리둘레', size='식전혈당(공복혈당)', sizes = (10,300))

#총 4개의 기준을 넣을수 있음 - 수축기혈압, 총콜레스테롤, 허리둘레, 식전혈당
# size로 각 node를 식전혈당으로 분별
```

### heatmap

```python
df = data.pivot_table(index='신장(5Cm단위)', columns = '체중(5Kg단위)', values = '허리둘레', fill_value = 0)
df
```

```python
plt.figure(figsize=(10,5))
plt.title("체중과 신장에 따른 허리둘레의 평균")
sns.heatmap(df, cmap='BuGn', annot = True)
```
