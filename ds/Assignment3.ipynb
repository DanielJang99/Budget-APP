# 3)

```python
query = "SELECT * FROM 'payments'"

payments = pd.read_sql(query, connect)

print(payments.shape)
payments.head()
```

```python
c1 = payments['연락처'].str.contains("^010|\+82")
payments.loc[~c1, '연락처']
```

```python
import numpy as np

def clean_phone_number(phone_number):
    if pd.isnull(phone_number):
        return pd.np.nan

#대시가 문자열에 이미 있다면 번호를 그대로 반환합니다.
    if "-" in phone_number:
        return phone_number
    
#대시가 문자열에 없다면 자릿수에 맞추어 대시를 넣어줍니다.
    else:
        phone_number = phone_number[0:3] + "-" + phone_number[3:7] + '-' + phone_number[7:]
        
        return phone_number
```

```python
c1_backup = c1.copy()

payments["연락처(clean)"] = payments.loc[c1]["연락처"].apply(clean_phone_number)
payments.head()
```

```python
payments['수업타입'] = payments['신청수업'].str.split(" ").str[0]
payments['기수'] = payments['신청수업'].str.split(" ").str[1]
cols = ['신청수업','수업타입','기수']
payments[cols].head()
```

# 4)

```python
payments['신청날짜'] = pd.to_datetime(payments['신청날짜'])
```

```python
def PaymentStatus(status):
    if status == "결제 완료":
        return True
    else:
        return False 
```

```python
payments['상태(bool)'] = payments['상태'].apply(PaymentStatus)
cols = ['상태','상태(bool)']
payments[cols].head()
```

# 5)

```python
payments['신청날짜(연)'] = payments['신청날짜'].dt.year
payments['신청날짜(월)'] = payments['신청날짜'].dt.month
payments['신청날짜(일)'] = payments['신청날짜'].dt.day
payments['신청날짜(시)'] = payments['신청날짜'].dt.hour
payments['신청날짜(분)'] = payments['신청날짜'].dt.minute
payments['신청날짜(초)'] = payments['신청날짜'].dt.second
payments['신청날짜(요일)'] = payments['신청날짜'].dt.day_name()
```

```python
payments['금액(int)'] = payments['금액'].str.replace(',','').astype(int)
```

```python
cols = ['신청날짜','신청날짜(연)','신청날짜(월)','신청날짜(일)','신청날짜(시)','신청날짜(분)','신청날짜(초)','신청날짜(요일)','금액','금액(int)']
payments[cols].head()
```

```python
PurchaseCompleted = payments[payments['상태'] == "결제 완료"] 
VIP_Email = PurchaseCompleted['이메일'].value_counts()[lambda x: x>2].index.tolist()
VIP_Email
```

```python
def checkVIP(data, list1):
    if data in list1:
        return True
    else:
        return False
PurchaseCompleted["isVIP"] = PurchaseCompleted['이메일'].apply(checkVIP, args = [VIP_Email])
VIPs = PurchaseCompleted[PurchaseCompleted["isVIP"]== True]
VIPs
```

```python
VIP_table = VIPs.pivot_table(index = ['이메일'], values= ['상태'], aggfunc = len)
VIP_table.sort_values("상태",ascending=False).head()
```
